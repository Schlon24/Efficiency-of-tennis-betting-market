### Effekt of the Favorite-Longshot Bias

### Last 10 games variable

### Return: Sum of return in the last 3 Games (Koning 2012)

```python
last_match_played = defaultdict(lambda: ["x",[0]])
def L3_fav_return(row):

    fav = row["favPlayer"]
    out = row["outPlayer"]

    favs = tuple(sorted([fav]))
    dogs = tuple(sorted([out]))

    result = sum(last_match_played[favs][1][-3:]) if (last_match_played[favs][0] == row["favPlayer"]) | (
        last_match_played[dogs][0] == row["favPlayer"]) else 0
    fav_return= 1/row["fav_p"]-1 if row["favPlayer"]==row["Winner"] else -1
    out_return=1/row["out_p"]-1 if row["outPlayer"]==row["Winner"] else -1

    last_match_played[favs][0] = row["favPlayer"]
    last_match_played[favs][1].append(fav_return)

    last_match_played[dogs][0]= row["outPlayer"]
    last_match_played[dogs][1].append(out_return)

    return result
```

```python
Original["L3_fav_return"]=Original.apply(L3_fav_return,axis=1)
Replication["L3_fav_return"]=Replication.apply(L3_fav_return,axis=1)
```

```python
# Für favorisierten Spieler Wechsel des Turnierort in km
last_match_played = defaultdict(lambda: ["x","y","z"])
def travel_km(row):

    fav = row["favPlayer"]
    out = row["outPlayer"]


    favs = tuple(sorted([fav]))
    dogs = tuple(sorted([out]))

    result = 1 if (last_match_played[favs][0] == row["favPlayer"]) | (
        last_match_played[dogs][0] == row["favPlayer"]) else 0
    travel = 1 if (result == 1) & (last_match_played[favs][1] != row["Location"])  else 0
    travel_km=  distance.distance(last_match_played[favs][2], row["coordinates"]).km if (travel == 1) else 0


    last_match_played[favs][0] = row["favPlayer"]
    last_match_played[favs][1]=(row["Location"])
    last_match_played[favs][2]=(row["coordinates"])
    last_match_played[dogs][0]= row["outPlayer"]
    last_match_played[dogs][1]=(row["Location"])
    last_match_played[dogs][2]=(row["coordinates"])
    
    return travel_km
```

```python
Original["travel_km"]=Original.apply(travel_km,axis=1)
Replication["travel_km"]=Replication.apply(travel_km,axis=1)
```

```python
# Für Außenseiter Wechsel des Turnierort in km
last_match_played = defaultdict(lambda: ["x","y","z"])
def travelOut_km(row):

    fav = row["favPlayer"]
    out = row["outPlayer"]


    favs = tuple(sorted([fav]))
    dogs = tuple(sorted([out]))

    result = 1 if (last_match_played[favs][0] == row["outPlayer"]) | (
        last_match_played[dogs][0] == row["outPlayer"]) else 0
    travel = 1 if (result == 1) & (last_match_played[dogs][1] != row["Location"])  else 0
    travel_km=  distance.distance(last_match_played[dogs][2], row["coordinates"]).km if (travel == 1) else 0
    
    last_match_played[favs][0] = row["favPlayer"]
    last_match_played[favs][1]=(row["Location"])
    last_match_played[favs][2]=(row["coordinates"])
    last_match_played[dogs][0]= row["outPlayer"]
    last_match_played[dogs][1]=(row["Location"])
    last_match_played[dogs][2]=(row["coordinates"])
    
    return travel_km
```

```python
Original["travelOut_km"]=Original.apply(travelOut_km,axis=1)
Replication["travelOut_km"]=Replication.apply(travelOut_km,axis=1)
```

### Logit Model

```python
Original["L3_dif_return"] = np.where(Original.Choice == 1, Original.L3_fav_return -
                                      Original.L3_out_return, Original.L3_out_return-Original.L3_fav_return)

Replication["L3_dif_return"] = np.where(Replication.Choice == 1, Replication.L3_fav_return -
                                      Replication.L3_out_return, Replication.L3_out_return-Replication.L3_fav_return)
```

```python
Original["wetteKMAdv"] = np.where(Original.Choice == 1, Original["travelOut_km"] -
                                  Original["travel_km"], Original["travel_km"]-Original["travelOut_km"])
Replication["wetteKMAdv"] = np.where(Replication.Choice == 1, Replication["travelOut_km"] -
                                     Replication["travel_km"], Replication["travel_km"]-Replication["travelOut_km"])
```

```python
X = sm.add_constant(x.T)
gam_bs = statsmodels.discrete.discrete_model.Logit(y,X)

res_bs = gam_bs.fit()

res_bs.summary()
```
